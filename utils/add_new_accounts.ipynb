```python
# Helper function to format list of new screen names to be added
# Creates a .txt file with one screen name on each line
# Required for twarc
def setup_twarc_input_file(screen_names):
    with open('./screen_names.txt', 'w') as f:
        for s in screen_names:
            f.write(s + '\n')
```

```python
# List of Twitter user names to be added
screen_names = ["KrishnaDharma","iskcon","radhanathswami","GaurangaDas","gaurgopald", "iskconchowpatty","dandavats_com","ISKCONPune", "SDGonline","IskconBbsr","HareKrsnaTV","Madhava_BBSR","bkgoswami","GKGmedia","JayapatakaSwami","india_iskcon","ISKCONnews"]
```

```python
setup_twarc_input_file(screen_names)
```

### Merge old and new files

```python
df_combo = pd.concat([df_new, df_old]).drop_duplicates('id_str').reset_index(drop=True).fillna('')
df_combo
```

### Write the combined file to the .json, .xlsx, and .csv files
